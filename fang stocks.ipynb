```python
import seaborn as sns
import pandas as pd

data_path = "fang stocks.csv"
data = pd.read_csv (data_path)

data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year
data



```

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

max_prices = data.groupby(['Year', 'Company'])['Close'].max().reset_index()
plt.figure(figsize=(20, 6))
for stock in max_prices['Company'].unique():
    stock_data = max_prices[max_prices['Company'] == stock]
    plt.plot(stock_data['Year'], stock_data['Close'], label=stock)

plt.xlabel('Year')
plt.ylabel('Max Closing Price')
plt.title('Max Closing Price of FANG Stocks per Year')
plt.legend()
plt.gca().set_facecolor('beige')
plt.show()


```
